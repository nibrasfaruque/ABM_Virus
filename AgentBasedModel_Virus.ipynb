```python
from virus_util import MakeGrid, RandomPerson, Iterate, Count
```

```python
GridSize = 40  
NumOfPeople = 100 #Initial not affected
NumOfInfect = 20  #Initial affected
daysInfect = 14 #Life Expectancy in days after infection

wGrid = MakeGrid(GridSize) #Creating the space using grid size 40x40 = 1600 spaces, each person can occupy a space 
RandomPerson(wGrid, NumOfPeople, 'U', 0) #Plotting uninfected people
RandomPerson(wGrid, NumOfInfect, 'I', daysInfect) #Plotting infected people 
```

```python
for step in range(90): #Iteration of 90 days to simulate the model
    Iterate(wGrid, daysInfect)
```

```python
Count(wGrid) #number of uninfected, infected, total of both
```

```python
#Scenario A: Catastrophic Outcome; 3 trials 
#1st Trial
aGrid = MakeGrid(10)  
RandomPerson(aGrid, 80, 'U', 0) 
RandomPerson(aGrid, 20, 'I', 10) 
```

```python
for step in range(100): 
    Iterate(aGrid, 2)
```

```python
#Conclusion: Overall in all 3 trials, the change went from mostly uninfected people to either only infected people or no one surviving. 
#This change happened because most of the space were occupied by a infected or not infected person, the close proximity to each person caused
#the spread of the virus to go through the entire population. Especially, since the step times are greater than than that of days expectancy after infection
#Trial 1 started with 80 uninfected and 20 infected in a 10x10 square space and resulted in no one surviving. 
#Trial 2 started with 250 uninfected and 40 infected in a 20x20 square space and result in only 20 survivors but 19 are infected with only 1 uninfected.
#Trial 3 started with 1250 uninfected and 300 infected in a 50x50 square space and resulted in only one survivor.
```
